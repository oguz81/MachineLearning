```python
## KNN method on Fashion MNIST dataset
# Oguz Demirtas

import numpy as np
from tensorflow import keras
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

(X_train, Y_train), (X_test, Y_test) = keras.datasets.fashion_mnist.load_data()

X_train_flatten = X_train.reshape(len(X_train), -1)
X_test_flatten = X_test.reshape(len(X_test), -1)

standard_scaler = StandardScaler()

X_train_flatten = X_train_flatten.astype('float32') / 255
X_test_flatten = X_test_flatten.astype('float32') / 255
X_train_scaled = standard_scaler.fit_transform(X_train_flatten)
X_test_scaled = standard_scaler.transform(X_test_flatten)

X_train, X_test, Y_train, Y_test = train_test_split(X_train_scaled, Y_train, test_size=0.2, random_state=42)
```

```python
accuracy_for_k = []
for i in range(1,5):
  knn = KNeighborsClassifier(n_neighbors=i)
  knn.fit(X_train, Y_train)
  prediction = knn.predict(X_test)
  accuracy_for_k.append(accuracy_score(Y_test, prediction))
  print("Accuracy for k =", i, ":", accuracy_for_k[i-1])

# plot the scores againist to neighbor number
import matplotlib.pyplot as plt
plt.plot(range(1,5), accuracy_for_k)
plt.xlabel("Number of neighbors")
plt.ylabel("Accuracy")
plt.show()
```

```python
accuracy_for_metrics = []
knn_metrics = ["euclidean", "manhattan", "minkowski", "chebyshev", "cosine", "hamming"]
for metric in knn_metrics:
  knn = KNeighborsClassifier(n_neighbors=1, metric=metric)
  knn.fit(X_train, Y_train)
  prediction = knn.predict(X_test)
  accuracy_for_metrics.append(accuracy_score(Y_test, prediction))
  print("Accuracy for metric =", metric, ":", accuracy_for_metrics[-1])

# plot the scores againist to neighbor number
plt.plot(knn_metrics, accuracy_for_metrics)
plt.xlabel("Metric (k = 1)")
plt.ylabel("Accuracy")
plt.show()
```
